```python
import torch
from pathlib import Path

def load_model(model, directory: str, model_name: str):
    # Ensure the directory path is valid
    directory = Path(directory)
    model_type = model_name.split('_')[0]
    model_load_path = directory / model_name
    
    # Check if the specified model file exists
    assert model_load_path.is_file(), f"[Error]: Model file {model_load_path} not found."

    # Load the model
    print(f"[Loading model from]: {model_load_path}")
    model.load_state_dict(torch.load(model_load_path)) 
    
    return model, model_type
```

```python
def test_step(model1, model2, data_loader, device, model1_type: str, model2_type: str, weight1: float, weight2: float):
    model1.eval()
    model2.eval()

    test_loss = 0

    preds_model1 = []
    preds_model2 = []

    true_values = []
    
    with torch.no_grad():
        for batch, (X, y) in enumerate(data_loader):
            X, y = X.to(device), y.to(device)
            true_values.append(y.cpu().numpy())

            # Get predictions from model1
            if model1_type == "LSTMVAE":
                _, y_pred1, _ = model1(X)
                y_pred1 = y_pred1 * weight1
                preds_model1.append(y_pred1.cpu().numpy())
            elif model1_type in ["SimpleLSTM", "Transformer", "Ensemble"]:
                y_pred1 = model1(X)  # Corrected variable name to y_pred1
                y_pred1 = y_pred1 * weight1
                preds_model1.append(y_pred1.cpu().numpy())

            # Get predictions from model2
            if model2_type == "LSTMVAE":
                _, y_pred2, _ = model2(X)
                y_pred2 = y_pred2 * weight2
                preds_model2.append(y_pred2.cpu().numpy())
            elif model2_type in ["SimpleLSTM", "Transformer", "Ensemble"]:
                y_pred2 = model2(X)
                y_pred2 = y_pred2 * weight2
                preds_model2.append(y_pred2.cpu().numpy())

    # Concatenate all predictions
    preds_model1 = np.concatenate(preds_model1, axis=0)
    preds_model2 = np.concatenate(preds_model2, axis=0)
    true_values = np.concatenate(true_values, axis=0)

    preds = (preds_model1 + preds_model2) / (weight1 + weight2)

    # Calculate the loss using the weighted predictions
    loss_function = nn.MSELoss()
    true_values_tensor = torch.tensor(true_values).to(device).float()  # Ensure it's a float tensor

    # Calculate the loss for model1's predictions
    loss = loss_function(torch.tensor(preds).to(device).float(), true_values_tensor)

    test_loss = loss.item()

    return test_loss, preds, true_values
```

```python
from torch.utils.data import ConcatDataset, DataLoader
combined_dataset = ConcatDataset([train_loader.dataset, val_loader.dataset, test_loader.dataset])
combined_loader = DataLoader(combined_dataset, batch_size=batch_size, shuffle=False)
```

```python
weights = sorted([np.exp(-i) for i in range(0,2)])
weight1 = weights[0]
weight2 = weights[1]
```

```python
test_loss, preds, true_values = test_step(model1 = model1, model2 = model2, 
                                                   data_loader = combined_loader, 
                                                   device = device, 
                                                   model1_type = model1_type, model2_type = model2_type, 
                                                   weight1 = weight1, weight2 = weight2)
```
